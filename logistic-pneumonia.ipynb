```python

import glob
Normal=glob.glob("/kaggle/input/datasets/muhammadrehan00/chest-xray-dataset/train/normal/*.jpg")
pneumonia=glob.glob("/kaggle/input/datasets/muhammadrehan00/chest-xray-dataset/train/pneumonia/*.jpg")
```

```python
print(len(Normal)+len(pneumonia))
```

# Training data feautures

```python
from PIL import Image
train_data=[]
train_label=[]
import numpy as np
for i in Normal:
    img=Image.open(i)
    img=img.convert('L')
    img=img.resize((100,100))
    arr=np.array(img)
    arr=arr.flatten()
    train_data.append(arr)
    train_label.append(0)

for i in pneumonia:
    img=Image.open(i)
    img=img.convert('L')
    img=img.resize((100,100))
    arr=np.array(img)
    arr=arr.flatten()
    train_data.append(arr)
    train_label.append(1)

print(type(train_data))
print(type(train_data[0]))
x_train=np.array(train_data)
y_train=np.array(train_label)
print(type(x_train))
print(type(x_train[0]))
```

# Test data feautures

```python
from PIL import Image
test_data=[]
test_label=[]
import numpy as np
for i in Normal_test:
    img=Image.open(i)
    img=img.convert('L')
    img=img.resize((100,100))
    arr=np.array(img)
    arr=arr.flatten()
    test_data.append(arr)
    test_label.append(0)

for i in pneumonia_test:
    img=Image.open(i)
    img=img.convert('L')
    img=img.resize((100,100))
    arr=np.array(img)
    arr=arr.flatten()
    test_data.append(arr)
    test_label.append(1)

print(type(test_data))
print(type(test_data[0]))
x_test=np.array(test_data)
y_test=np.array(test_label)
print(type(x_test))
print(type(x_test[0]))

```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import numpy as np

x_train = np.load("X_train.npy")
y_train = np.load("y_train.npy")
x_test = np.load("X_test.npy")
y_test = np.load("y_test.npy")
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=7)

x_train_pca = pca.fit_transform(x_train_scaled)
x_test_pca = pca.transform(x_test_scaled)

print(x_train_pca.shape)
```

```python
model = LogisticRegression(max_iter=1000,C=0.001)

model.fit(x_train_pca, y_train)
```

```python
y_pred = model.predict(x_test_pca)
```

# TESTING DATA ACCURACY

```python


print("Accuracy:", accuracy_score(y_test, y_pred))
```

# TRAINING DATA ACCURACY

```python
print("Accuracy:", model.score(x_train_pca, y_train))
```
